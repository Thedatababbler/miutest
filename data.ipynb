```python
import os
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

class CustomDataset(Dataset):
    def __init__(self, image_path, label_path, transform=None):
        self.image_path = image_path
        self.label_path = label_path
        self.transform = transform
        self.images = os.listdir(image_path)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        # Get the file name and construct the full path for image and mask
        img_name = self.images[idx]
        img_full_path = os.path.join(self.image_path, img_name)
        mask_full_path = os.path.join(self.label_path, img_name)

        # Read the image and mask
        image = Image.open(img_full_path).convert("RGB")
        mask = Image.open(mask_full_path).convert("L")  # Assuming mask is grayscale

        # Apply transformations if any
        if self.transform is not None:
            image = self.transform(image)
            mask = self.transform(mask)

        return image, mask, img_full_path

# Define transformations if needed, for example resizing images and converting them to PyTorch tensors
transformations = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Create the dataset
image_path = 'DATA/COVID-19_Radiography_Dataset/Normal/images'
label_path = 'DATA/COVID-19_Radiography_Dataset/Normal/masks'
dataset = CustomDataset(image_path=image_path, label_path='/label', transform=transformations)

# Create the dataloader
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# Now you can iterate through the dataloader
for images, masks in dataloader:
    # Here, 'images' is a batch of images and 'masks' is the corresponding batch of masks
    # Implement your training loop or data handling here
    print()
```
